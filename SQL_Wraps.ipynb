What is the subscriber_type for most popular start station and is the most popular subscriber type also popular in most popular start station?


```python
%%bigquery --project deep-presence-338719
SELECT COUNT(start_station_name) as count, start_station_name as name, subscriber_type as subs
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY name, subs
ORDER BY count DESC
```

```python
%%bigquery --project deep-presence-338719
SELECT COUNT(subscriber_type) as count,subscriber_type as subs
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY subs
ORDER BY count DESC
```
