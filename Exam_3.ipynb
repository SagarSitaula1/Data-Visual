Let me add a new column to the dataset to keep track of the total number of goals and assists scored by all Manchester United players in each row.

First, I'll write a function that takes a row of data and returns the sum of the goals and assists for that row. Then I'll use the str.contains function to extract only the entries from the Club Column that contain Manchester United. Finally, I'll utilize Lambda to feed those data to our function, which will return the sum and save it in a new column named Manchester United.

```python
def ManchesterUnitedGoals(row):
  return row.Goals

def ManchesterUnitedAssists(row):
  return row.Assists
```

```python
df2 = df[df['club'].str.contains('Manchester United')]

df['MU_goals'] = df2.apply(lambda row: ManchesterUnitedGoals(row), axis=1)

df['MU_assists'] = df2.apply(lambda row: ManchesterUnitedAssists(row), axis=1)
```

```python
df['MU_goals'] = df['MU_goals'].fillna(0).astype(int)
df['MU_goals'] = df['MU_goals'].astype('int')
df['MU_assists'] = df['MU_assists'].fillna(0).astype(int)
df['MU_assists'] = df['MU_assists'].astype('int')
df[df['club'].str.contains('Manchester United')]
```
